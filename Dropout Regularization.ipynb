```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
import warnings
warnings.filterwarnings("ignore") #is used in Python to suppress warnings that might otherwise be displayed during the execution of a script.
#This line tells Python to ignore all warnings that would normally be issued.
```

```python
df = pd.read_csv('sonar_dataset.csv', header=None)
df.sample(10)
```

```python
df.shape
```

```python
#chek for nan values on every feature
df.isna().sum()
```

```python
df.columns
```

```python
#on column number 60....which will be our Y, check the count of unique values
df[60].value_counts()
```

```python
#Describing our X and y features
X = df.drop(60, axis=1)
y = df[60]
y.sample(10)
```

we need to classify true or false....1 or 0 for the y feature

```python
#Describing y again
y = pd.get_dummies(y, drop_first=True) #in this line...in the column y...we are taking the unique values .i.e R and M and puting them in separate columns....in which the value of rows will be True(1) and not True(0). Then drop_first drops row number 2 (one-hot encoding)
y = y.astype(int)

y.sample(5) # R --> 1 and M --> 0
```

```python
y.value_counts()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

Using Deep Learning Model

Model without Dropout Layer

```python
model = keras.Sequential([
    keras.layers.Dense(60, activation='relu', input_shape=(60,)), #has 60 features...one dim
    keras.layers.Dense(30, activation='relu'), #we are passing the values mathematically to the relu function
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8) #batch_size=8: This defines the number of samples that will be propagated through the network at one time. The dataset is divided into batches of 8 samples, and the model's weights are updated after each batch.
```

```python
model.evaluate(X_test, y_test)
```

Training Accuracy >>> Test Accuracy

```python
y_pred = model.predict(X_test).reshape(-1) #reshape going towards the right
print(y_pred[:10])

#round the values to the nearest integer .i.e 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10] #true values
```

```python
print(classification_report(y_test, y_pred))
```

Model with Dropout Layer

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, activation='relu', input_dim=60),
    keras.layers.Dropout(0.5), #a fraction of the neurons are dropped out
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

Training Accuracy is still good but Test Accuracy Improved

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
print(classification_report(y_test, y_pred))
```

You can see that by using dropout layer test accuracy increased from 0.75 to 0.83
